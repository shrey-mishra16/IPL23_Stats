```python
import pandas as pd
```

```python
df = pd.read_csv("combined_csv.csv")
```

```python
df.info()
```

```python
unique_batters = df['batter'].unique()

# Create a dictionary to store each batter's DataFrame
batter_data_frames = {}

for batter in unique_batters:
    # Filter the data by each batter
    batter_data = df[df['batter'] == batter]
    
    # Drop the 'comment' column
    batter_data = batter_data.drop(columns=['comment','batter','ballnumber'])
    
    # Store the batter's DataFrame in the dictionary
    batter_data_frames[batter] = batter_data
```

```python
unique_bowlers = df['bowler'].unique()

# Create a dictionary to store each batter's DataFrame
bowler_data_frames = {}

for bowler in unique_bowlers:
    # Filter the data by each batter
    bowler_data = df[df['bowler'] == bowler]
    
    # Drop the 'comment' column
    bowler_data = bowler_data.drop(columns=['comment','bowler','ballnumber'])
    
    # Store the batter's DataFrame in the dictionary
    bowler_data_frames[bowler] = bowler_data
```

```python
unique_outcomes = df['outcome'].unique()

# Print the unique outcomes
print(unique_outcomes)
```

```python
specific_outcomes = [0 ,4,6 ,5 ,7]
filtered_data = df[df['outcome'].isin(specific_outcomes)]

# Create a cross-tabulation table for batter and bowler
cross_match = pd.crosstab(index=filtered_data['batter'], columns=filtered_data['bowler'], values=filtered_data['outcome'], aggfunc='count', dropna=True)

# Fill the NaN values with 0
cross_match = cross_match.fillna(0)

print(cross_match)
```

```python
max_cross_match = cross_match.stack().idxmax()
max_total = cross_match.values.max()

print(f"The pair with the maximum cross-match is ({max_cross_match[0]}, {max_cross_match[1]}) with a total count of {max_total}.")
```

```python
cross_match = pd.crosstab(index=filtered_data['batter'], columns=filtered_data['bowler'], values=filtered_data['outcome'], aggfunc='count', dropna=True)

# Fill the NaN values with 0
cross_match = cross_match.fillna(0)

# Filter the cross-match table based on count values
filtered_cross_match = cross_match[cross_match > 2].stack().reset_index()
filtered_cross_match.columns = ['Batter', 'Bowler', 'Count']

# Iterate through the filtered cross-match table and print the results
for index, row in filtered_cross_match.iterrows():
    print(f"Pair ({row['Batter']}, {row['Bowler']}) has a cross-match count of {row['Count']}.")
```

```python
sorted_cross_match = cross_match.stack().sort_values(ascending=False)
second_highest_cross_match = sorted_cross_match.iloc[1]
second_highest_pair = sorted_cross_match.index[1]

print(f"The pair with the second-highest cross-match is {second_highest_pair} with a total count of {second_highest_cross_match}.")
```

```python
# Replace 'w' with 12 and remove all other letters
df['outcome'] = df['outcome'].replace({'w': 12, 'wd': 0}, regex=True).map(lambda x: int(''.join(filter(str.isdigit, str(x)))))

print(df)
```

```python
df = df.drop(df[df['batter'] == 'n Sundar'].index)
```

```python
import matplotlib.pyplot as plt
```

```python
def bowlerstate(x):
    filtered_data = df[(df['outcome'] != 12) & (df['bowler'] == x)]

    # Group the data by batter and calculate the sum of outcomes
    sum_outcomes = filtered_data.groupby('batter')['outcome'].sum()

    # Create a bar chart of the sum of outcomes for each batter
    sum_outcomes.plot(kind='bar')

    # Set the title and axis labels
    plt.title('Outcomes for'+x+'against All Batters')
    plt.xlabel('Batter')
    plt.ylabel('Sum of Outcomes (excluding wickets)')

    # Display the chart
    plt.show()
```

```python
x = input("bowler name")
bowlerstate(x)
```

```python
def batterstate(x):
    filtered_data = df[(df['outcome'] != 12) & (df['batter'] == x)]

    # Group the data by batter and calculate the sum of outcomes
    sum_outcomes = filtered_data.groupby('bowler')['outcome'].sum()

    # Create a bar chart of the sum of outcomes for each batter
    sum_outcomes.plot(kind='bar')

    # Set the title and axis labels
    plt.title('Outcomes for'+x+'against All Bowlers')
    plt.xlabel('Bowler')
    plt.ylabel('Sum of Outcomes (excluding wickets)')

    # Display the chart
    plt.show()
```

```python
y = input("batter name")
batterstate(y)
```

```python
filtered_data = df[df['outcome'] != 12]

# Group the data by batter and bowler and calculate the sum of outcomes (excluding 12)
sum_outcomes = filtered_data.groupby(['batter', 'bowler'])['outcome'].sum()

# Get the index (i.e., the bowler) with the highest sum of outcomes (excluding 12) for each batter
highest_outcome_bowler = sum_outcomes.groupby('batter').idxmax().reset_index()

# Create a dot chart of the bowler with the highest sum of outcomes (excluding 12) for each batter
plt.scatter(highest_outcome_bowler['batter'], highest_outcome_bowler['bowler'])

# Set the title and axis labels
plt.title('Bowler with Highest Total Outcome (excluding 12) for Each Batter')
plt.xlabel('Batter')
plt.ylabel('Bowler')

# Display the chart
plt.show()
```
